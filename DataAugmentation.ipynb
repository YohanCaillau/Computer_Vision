```python
X_reshape = np.reshape(X_array, (X_array.shape[0],X_array.shape[1],X_array.shape[2], 1))
X_reshape.shape
```

```python
from matplotlib import pyplot
# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(X_array[i], cmap=pyplot.get_cmap('gray'))
# show the plot
pyplot.show()
```

```python
from numpy import expand_dims
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot

# create image data augmentation generator
datagen = ImageDataGenerator(zca_whitening=True,
    featurewise_std_normalization=True,
    rotation_range=90,
    width_shift_range=0.3,
    height_shift_range=0.3)

model = datagen.fit(X_reshape)

for X_batch in datagen.flow(X_reshape, batch_size=9):
    # create a grid of 3x3 images
    for i in range(0, 9):
        pyplot.subplot(330 + 1 + i)
        pyplot.imshow(X_batch[i].reshape(64, 64), cmap=pyplot.get_cmap('gray'))
        # show the plot
    pyplot.show()
    break
```
